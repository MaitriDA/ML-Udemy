# SVM - Support Vector Machine

There can be many ways to classify the data points. For the below given data points, we can divide them by drawing a vertical line, a horizontal line, slant line etc.

In SVM the best bit line is searched through maximum margin. The blue line is drawn equi-distance from both read and green point. The points from each zone which are used to draw the line is known as support Vector. The other points do not contribute to the algorithm.

## STEP 1 - Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

## STEP 2 - Training the model

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',random_state=0)
svm.fit(x_train,y_train)
```
