# Logistic Regression

Logistic Regression is generally used when the end result is Yes or No.

We convert linear regression to logistic regression using the following formula. Blue box is linear regressiona dn green box is the logistic regression.

We can predict the probability. For the person whe is 20 years old the probability of accepting the offer is 0.7%, the person with 30 years of age 23% and so on. So by this way we can predict the probability using Logistic Regression fir line.

Instead of probability we can predict the result as yes or no. Weather the person will accept the offer or not. If the probability is less than 50% he/she will not accept the offer else he/she will accept the offer.

## Step 1 - Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

## Step 2 - Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

## STEP 3 - Training the model 

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
```

## STEP 4 - Predict the result

```python
lr.predict(sc.transform([[30,10000]]))
```

```python
y_pred=lr.predict(x_test)
np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
```

#### Confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
lr.score(x_test,y_test)
```

## STEP 5 - Visualisation

#### Visualisation the training set 

```python
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = x_set[:, 1].min() - 10, stop = x_set[:, 1].max() + 10, step = 0.25))
plt.contourf(x1, x2, lr.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

#### Visualisation the test set 

```python
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_test), y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = x_set[:, 1].min() - 10, stop = x_set[:, 1].max() + 10, step = 0.25))
plt.contourf(x1, x2, lr.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Testing set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
