# Simple Linear Regression

y = b0 + b1*x<br>
Here y is the dependent Variable and x in the independent variable<br>
b0 is the constant and b1 is the coefficient<br>
<br>
Taking the example of salary and experience. Salary here acts as dependent variable and experiene is independent variable.<br>
salary = b0 + b1*experience.<br>
<br>
The login behind simple linear regression is to find the best fir line based on the independent and dependent variables


b0 the constant is the y intercept. So it means the salary of the perso with 0 experience.<br>
b1 is the slop of the line

## How to get the best fit line 

In the figure red represents the actual data and green is the data predicted by the model.

So in linear regression first all possible lines are drawn.<br>
Then using the formula SUM (y - y^)^2 is calculated for all the lines.<br>
The line that gives the minimum sum is the best fit line.<br>
This is know as Ordinary Least Square method.

## STEP 1 -  Data Pre-processing 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('Salary_Data.csv')
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

## STEP 2 -  Building the model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

## STEP 3 - Predicting the test set results 

```python
y_predict=lr.predict(x_test)
#Predicted Salary
print(y_predict)
```

```python
#Actual Salary
y_test
```

```python
lr.score(x_test,y_test)
```

## STEP 4 -  Visualisation 

```python
plt.scatter(x_train,y_train,color='red',marker='+')
plt.plot(x_train,lr.predict(x_train),color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red',marker='+')
plt.plot(x_train,lr.predict(x_train),color='blue')
plt.title('Salary vs Experience (Testing set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_predict,color='red',marker='+')
plt.plot(x_train,lr.predict(x_train),color='blue')
plt.title('Salary vs Experience (Predicted values)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

From the above graph we can observe graph we can observe that the predicted values are always on the linear regression line
