# Polynomial Regression

It is very similar to multiple regression but it has only one independent variable. And the x1 has powers of 2,3...etc<br>
y = b0 + b1*x + b2*x^2 + b3+x^3 + ... + bnx^n 

## STEP 1 - Data Pre-Processing 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Position_Salaries.csv')
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

## STEP 2 - Training the model 

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

Polynomial Regression can as treated as multiple regression as well, but with a slight modification.<br>
In multiple regression we had x1,x2,x3,x4,....xn<br>
In polynomial regression we have x1,x1^2,x1^3,x1^4.....x1^n . This is known as power features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#try with degree 1 (It will be same as linear regression)
pf=PolynomialFeatures(degree=6)
x_poly=pf.fit_transform(x)
pr=LinearRegression()
pr.fit(x_poly,y)
```

In tne above code we make a power feature matrix which is x_poly adn the use the technique of multiple regression

## STEP 3 - Visualisation  

```python
plt.scatter(x,y,color='red',marker='+')
plt.plot(x,lr.predict(x),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red',marker='+')
plt.plot(x,pr.predict(x_poly),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

### Smooth graph 

```python
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color='red',marker='+')
plt.plot(x_grid,pr.predict(pf.fit_transform(x_grid)),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

## STEP 4 - Predicting the result 

```python
#linear regression
lr.predict([[6.5]])
```

```python
#polymonial regression
pr.predict(pf.fit_transform([[6.5]]))
```
