# Multiple Regression

Multiple regression is as linear regression but it has more than one independent variables.<br>
LINEAR REGRESSION: y = b0 + b1*x<br>
MULTIPLE REGRESSION: y= b0 + b1*x1 + b2*x2 + ..... + bn*xn
