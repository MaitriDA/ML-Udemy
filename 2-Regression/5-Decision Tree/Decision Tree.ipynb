# Decision Tree

There are two types of trees in ML. One is Classification tree and the other is Regression tree. This part is about Regression tree

The scratter plot represents the data. It has two independent variables x1 and x2 and we are predicting the third dependent variable y.<br>
On applying the Decision tree Algorithm, the plot is divided into segments.<br>
The split are made on the basis of informatio entropy.<br>
Each segment is called leaf. The final leaves are called terminal leaves

lets take an example of x1=30 and x2=50. So the algorithm traverse the tree based on values of x1 and x2 and then predicts the value of y. One the correct segment (leaf) is reached then the average of the points inside that leaf is taken and is the predicted value.

for x1=30 and x2=50 the predicted value if y would be -64.1.

## STEP 1 - Data Pre - Processing 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Position_Salaries.csv')
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

## STEP 2 - Training the model on whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor(random_state=0)
dtr.fit(x,y)
```

### Predicting the result 

```python
dtr.predict([[7.5]])
```

## STEP 3 - Visualisation

```python
x_grid=np.arange(min(x),max(x),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color='red')
plt.plot(x_grid,dtr.predict(x_grid))
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
